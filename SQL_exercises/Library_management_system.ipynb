```python
import pyodbc 
import pandas as pd
conn = pyodbc.connect('Driver={SQL Server};'
                     'Server=LAPTOP-TOJSOVMJ\SQLEXPRESS;'
                     'Database=library_management;'
                     'Trusted_Connection=yes;')
```

```python
def showinfobook():
    cursor = conn.cursor()
    cursor.execute('SELECT * FROM sach_info')
    rows = cursor.fetchall()
    header = 'Mã Sách'.ljust(10, ' ') + 'Tên Sách'.ljust(40,' ') + 'Mã TL'.ljust(10, ' ') +'Mã NXB'.ljust(10, ' ') +'Năm XB'.ljust(10, ' ')+ 'Số Trang'.ljust(10, ' ') + 'Giá tiền'.ljust(15, ' ')
    header_ = '\n'.ljust(100, '-')
    print(header,header_)
    for i in rows:
        row = i[0].ljust(10,' ') + i[1].ljust(40,' ') + i[2].ljust(10,' ') + i[3].ljust(10,' ') + i[4].ljust(10,' ')+ \
        i[5].ljust(10,' ') + str(i[6]).ljust(10,' ')
        print(row)
            
```

```python
def addbookinfo(masach, tensach ,matl, manxb, namxb, sotrang, giatien):
    cursor = conn.cursor()
    insert_sql = """INSERT INTO sach_info VALUES(?,?,?,?,?,?,?)"""
    cursor.execute(insert_sql,masach,tensach,matl, manxb,namxb, sotrang, giatien)
    cursor.commit()
```

```python
def updatebookinfo(masach, tensach ,matl, manxb, namxb, sotrang, giatien):
    update_sql = """UPDATE sach_info SET Ten_sach = ?, Ma_theloai = ?, Ma_nxb = ?, Nam_xb = ?, Sotrang = ?, Gia_tien = ? WHERE Ma_sach = ?"""
    cursor = conn.cursor()
    cursor.execute(update_sql, tensach ,matl, manxb, namxb, sotrang, giatien, masach)
    cursor.commit()
def deletebookinfo(masach):
    validation_sql = 'SELECT * FROM sach_info WHERE Ma_sach = ?'
    cursor = conn.cursor()
    cursor.execute(validation_sql, masach)
    result = cursor.fetchone()
    T = False
    if result is None:
        print(f'Mã sách = {masach} không tồn tại. Nhập lại để cập nhập.')
    else:
        confirm = input(f'bạn chắc chắn muốn xóa thông tin sách {result[1]} (y/n)')
        if confirm.lower()=='y':
            delete_sql = 'DELETE FROM sach_info WHERE Ma_sach = ?'
            cursor.execute(delete_sql, masach)
            conn.commit()
            print('Delete successfully')
            T = True
    return T
def findoutbookinfo(masach):
    validation_sql = 'SELECT * FROM sach_info WHERE Ma_sach = ?'
    cursor = conn.cursor()
    cursor.execute(validation_sql, masach)
    result = cursor.fetchone()
    cursor.commit()
    return result 
```
